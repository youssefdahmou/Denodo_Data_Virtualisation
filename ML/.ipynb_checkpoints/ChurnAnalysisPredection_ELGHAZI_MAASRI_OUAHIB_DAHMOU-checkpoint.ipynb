<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Définir les colonnes numériques et les colonnes categoriales</div>

```python
numerics_columns = ['tenure', 'MonthlyCharges', 'TotalCharges']
cate_columns = [col for col in df.columns if col not in numerics_columns]
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">

```python
categorical_columns = ["gender", "Partner", "Dependents", "PhoneService", "MultipleLines", "InternetService", "OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies", "Contract", "PaperlessBilling", "PaymentMethod"]
numerical_columns = ["SeniorCitizen", "tenure", "MonthlyCharges", "TotalCharges"]
```

```python
train_data, test_data = spark_df.randomSplit([0.8, 0.2], seed=42)
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Définir la fonction de hyperparameter tuning</div>

```python
def tune_model(model, param_grid, train_data, test_data):
    evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")  
    # Créer un cross-validator
    cv = CrossValidator(estimator=model, estimatorParamMaps=param_grid, evaluator=evaluator, numFolds=3)
    # Performer le grid search pour trouver les meilleurs parameters
    cv_model = cv.fit(train_data)
    # retourner le meilleur modèle
    best_model = cv_model.bestModel
    # Printer les meilleurs params
    print("Best Parameters:")
    for param, value in best_model.extractParamMap().items():
        print(f"{param.name}: {value}")
    # Evaluer sur train data
    train_predictions = best_model.transform(train_data)
    train_accuracy = evaluator.evaluate(train_predictions)
    print(f"Train Accuracy: {train_accuracy}")
    # Evaluer sur test data
    test_predictions = best_model.transform(test_data)
    test_accuracy = evaluator.evaluate(test_predictions)
    print(f"Test Accuracy: {test_accuracy}")  
    return best_model
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Définir les modèles</div>

```python
lr_model = LogisticRegression(labelCol="label", featuresCol="features")
dt_model = DecisionTreeClassifier(labelCol="label", featuresCol="features")
rf_model = RandomForestClassifier(labelCol="label", featuresCol="features")
gbt_model = GBTClassifier(labelCol="label", featuresCol="features")
nb_model = NaiveBayes(labelCol="label", featuresCol="features")
layers = [len(feature_columns), 100, 50, 2]  
ann_model = MultilayerPerceptronClassifier(labelCol="label", featuresCol="features", layers=layers, seed=42)
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Définir les hyperparameter de grid</div>

```python
lr_param_grid = ParamGridBuilder() \
    .addGrid(lr_model.regParam, [0.01, 0.1, 1.0]) \
    .addGrid(lr_model.elasticNetParam, [0.0, 0.5, 1.0]) \
    .build()

dt_param_grid = ParamGridBuilder() \
    .addGrid(dt_model.maxDepth, [5, 10, 15]) \
    .addGrid(dt_model.minInstancesPerNode, [1, 5, 10]) \
    .build()

rf_param_grid = ParamGridBuilder() \
    .addGrid(rf_model.maxDepth, [5, 10, 15]) \
    .addGrid(rf_model.minInstancesPerNode, [1, 5, 10]) \
    .addGrid(rf_model.numTrees, [10, 20, 30]) \
    .build()

gbt_param_grid = ParamGridBuilder() \
    .addGrid(gbt_model.maxDepth, [5, 10, 15]) \
    .addGrid(gbt_model.minInstancesPerNode, [1, 5, 10]) \
    .addGrid(gbt_model.maxIter, [10, 20, 30]) \
    .build()

nb_param_grid = ParamGridBuilder() \
    .addGrid(nb_model.smoothing, [0.0, 0.2, 0.4]) \
    .build()

ann_param_grid = ParamGridBuilder() \
    .addGrid(ann_model.maxIter, [100, 200, 300]) \
    .addGrid(ann_model.stepSize, [0.01, 0.1, 0.2]) \
    .build()
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Performer l'hyperparameter tuning</div>

```python
best_lr_model = tune_model(lr_model, lr_param_grid, train_data,test_data)
```

```python
best_dt_model = tune_model(dt_model, dt_param_grid, train_data,test_data)
```

```python
best_rf_model = tune_model(rf_model, rf_param_grid, train_data,test_data)
```

```python
best_gbt_model = tune_model(gbt_model, gbt_param_grid, train_data,test_data)
```

```python
best_nb_model = tune_model(nb_model, nb_param_grid, train_data,test_data)
```

```python
best_ann_model = tune_model(ann_model, ann_param_grid, train_data,test_data)
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Fonction pour calculer et stocker les métriques dans une dataframe</div>

```python
def evaluate_model(model_name, model_results):
    evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction")

    accuracy = evaluator.evaluate(model_results, {evaluator.metricName: "accuracy"})
    jaccard = evaluator.evaluate(model_results, {evaluator.metricName: "weightedTruePositiveRate"})
    f1_score = evaluator.evaluate(model_results, {evaluator.metricName: "weightedFMeasure"})
    recall = evaluator.evaluate(model_results, {evaluator.metricName: "weightedRecall"})

    return (model_name, accuracy, jaccard, f1_score, recall)
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
evaluer et stocker les métriques pour chaque modèle</div>

```python
metrics_list = []
for model_name, model_result in test_results.items():
    metrics_list.append(evaluate_model(model_name, model_result))
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Indexer , assembler les features en un vecteur et effectuer le label encoding </div>

```python
# Indexer les colonnes  categorielles
indexers = [StringIndexer(inputCol=col, outputCol=f"{col}_index").fit(df_pip) for col in categorical_columns]
# Assembler les features en un vecteur
feature_columns = [f"{col}_index" for col in categorical_columns] + numerical_columns
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
# Label encoding pour la colonne cibl 'Churn'
label_indexer = StringIndexer(inputCol="Churn", outputCol="label").fit(df_pip)
```

<div style="background-color:orangered; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Combiner tous les stages dans la PIPELINE </div>

```python
pipeline = Pipeline(stages=indexers + [label_indexer, assembler])
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Convertir en pandas dataframe pour faire SMOTE et séparer les features avec la variable cible </div>

```python
df_pandas = df_pip_transformed.toPandas()
X = df_pandas[feature_columns]
y = df_pandas["label"]
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Smote:</div>

<div style="text-align:center;height:70;border: 5px solid black;border-radius:20px;;border-radius:20px;;">
    <img src="https://rikunert.com/wp-content/uploads/2017/11/the-basic-principle-of-the-synthetic-minority-oversample-technique-smote-algorithm-5452514.png" width="700" alt="image1">
</div>

<div style="background-color:black; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
L'idée derrière SMOTE est de créer des exemples synthétiques de la classe minoritaire plutôt que de simplement répéter les exemples existants. Cela permet d'augmenter la taille de la classe minoritaire et d'améliorer ainsi la capacité du modèle à généraliser correctement.</div>

```python
oversampler = SMOTE(sampling_strategy=1.0, random_state=42)
X_resampled, y_resampled = oversampler.fit_resample(X, y)
```

```python
X_resampled.shape,y_resampled.shape
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Combiner les features oversamplé avec la variable cible en une dataframe </div>

```python
df_pip_resampled = pd.DataFrame(X_resampled, columns=feature_columns)
df_pip_resampled["label"] = y_resampled
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Convertir dans l'autre sense vers spark Dataframe</div>

```python
df_pip_resampled_spark = spark_pip.createDataFrame(df_pip_resampled)
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Deviser le jeu de données en train et test </div>

```python
train_data, test_data = df_pip_resampled_spark.randomSplit([0.8, 0.2], seed=42)
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Definir le modèle Random Forest </div>

```python
rf_model_pip = RandomForestClassifier(labelCol="label", featuresCol="features")
```

<div style="background-color:orangered; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Inclure le modèle Random forest dans PIPELINE </div>

```python
pipeline_rf = Pipeline(stages=indexers + [label_indexer, assembler, rf_model_pip])
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Définir les grids d'hyperparameters </div>

```python
rf_param_grid = ParamGridBuilder() \
    .addGrid(rf_model_pip.maxDepth, [5, 10, 15]) \
    .addGrid(rf_model_pip.minInstancesPerNode, [1, 5, 10]) \
    .addGrid(rf_model_pip.numTrees, [10, 20, 30]) \
    .build()
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Performer le tuning des hyperparametres </div>

```python
cv_evaluator = BinaryClassificationEvaluator(rawPredictionCol="rawPrediction", labelCol="label")
cv = CrossValidator(estimator=pipeline_rf, estimatorParamMaps=rf_param_grid, evaluator=cv_evaluator, numFolds=3)
cv_model = cv.fit(train_data)
best_rf_model_pip = cv_model.bestModel
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Afficher les parametre de modèle best_rf_pip </div>

```python
print("Best Random Forest Model Parameters:")
print(best_rf_model_pip.stages[-1].extractParamMap())
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Afficher l'accuracy sur l'ensemble de train</div>

```python
train_predictions = best_rf_model_pip.transform(train_data)
train_evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
train_accuracy = train_evaluator.evaluate(train_predictions)
print(f"Train Accuracy: {train_accuracy:.4f}")
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Afficher l'accuracy sur l'ensemble de test</div>

```python
test_predictions = best_rf_model_pip.transform(test_data)
test_evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
test_accuracy = test_evaluator.evaluate(test_predictions)
print(f"Test Accuracy: {test_accuracy:.4f}")
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
ploter la matrice de confusion </div>

```python
y_true = test_predictions.select("label").rdd.map(lambda x: x[0])
y_pred = test_predictions.select("prediction").rdd.map(lambda x: x[0])

confusion_matrix = MulticlassMetrics(test_predictions.select("prediction", "label").rdd)
conf_matrix = confusion_matrix.confusionMatrix().toArray()
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt=".0f", cmap="Blues", cbar=False)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Afficher le rapport de classification</div>

```python
report = classification_report(y_true.collect(), y_pred.collect(), target_names=["No Churn", "Churn"])
print("Classification Report:\n", report)
```

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Afficher la valeur de AUC</div>

```python
binary_evaluator = BinaryClassificationEvaluator(labelCol="label", rawPredictionCol="rawPrediction")
auc_score = binary_evaluator.evaluate(test_predictions, {binary_evaluator.metricName: "areaUnderROC"})
print(f"AUC Score: {auc_score:.4f}")
```

<div style="background-color:black; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Un AUC de 0.9399 est considéré comme très bon. Cela signifie que le modèle a une excellente capacité à distinguer entre les classes positives et négatives. En d'autres termes, la probabilité qu'un exemple positif soit classé avec un score plus élevé que celui d'un exemple négatif est de 93,99%, ce qui est très prometteur.</div>

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Visualiser la courbe ROC </div>

```python
test_predictions = best_rf_model_pip.transform(test_data)
y_true = test_predictions.select("label").rdd.map(lambda x: x[0])
y_scores = test_predictions.select("probability").rdd.map(lambda x: x[0][1])

fpr, tpr, _ = roc_curve(y_true.collect(), y_scores.collect())

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(auc(fpr, tpr)))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

<div style="background-color:black; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
 Un AUC élevé indique que la courbe ROC s'éloigne fortement de la ligne diagonale, ce qui suggère une excellente capacité discriminatoire du modèle.</div>

<div style="background-color:olive; color:white; font-size:15px; font-family:Verdana; padding:10px;font-weight:bold;border: 5px solid black;border-radius:20px;;border-radius:20px;;text-align:center">
Enfin Sauvgarder le modèle</div>
